```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
print("Libraries imported successfully!")
```

```python
iris_dataset = sns.load_dataset("iris")
iris_dataset.to_csv("iris_dataset.csv", index=False)
```

```python
print("✅ iris_dataset.csv saved successfully")
```

```python
iris = pd.read_csv("iris_dataset.csv")
```

```python
print("Shape:", iris.shape)
```

```python
print(iris.columns)
```

```python
print(iris.head())
```

```python
iris.describe()
```

```python
iris.info()
```

```python
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species")
plt.title("Sepal Length vs Petal Length by Species")
plt.show()
```

```python
iris.hist(bins=15)
plt.suptitle("Distribution of Iris Dataset Features")
plt.show()
```

```python
sns.boxplot(data=iris)
plt.title("Box Plot of Iris Features")
plt.show()
```

```python
sns.boxplot(data=iris, x="species", y="petal_length")
plt.title("Petal Length by Species")
plt.show()
```

```python
sns.boxplot(data=iris,x="species",y="sepal_length")
plt.title("sepal Length by Species")
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set figure size and style
plt.figure(figsize=(16, 10))
features = [ "sepal_width", "petal_width"]

# Loop through features to create subplots
for i, feature in enumerate(features):
    plt.subplot(1, 2,i+1)  # 2 rows, 2 columns
    sns.boxplot(data=iris, x="species", y=feature)
    plt.title(f"{feature.replace('_', ' ').title()} by Species")

plt.tight_layout()
plt.show()
```

## 🔍 Conclusion

- Petal features are more useful than sepal features for distinguishing species.
- Setosa is clearly different from the other two species.
- Outliers exist in sepal width, especially for Setosa.
- The dataset is clean, balanced, and ready for classification models.
